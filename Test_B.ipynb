```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'lxml')

table = soup.find('table')

table_row = table.find_all('tr')

print(table)
```

```python
import csv
csv_file=open('toronto.csv','w')
csv_writer=csv.writer(csv_file)
csv_writer.writerow(['Postcode', 'Borough', 'Neighbourhood'])
for tr in soup.find_all('tr')[1:]:
    tds = tr.find_all('td')
    if len(tds)==3:
        Postcode=tds[0].text
        Borough=tds[1].text
        Neighbourhood=tds[2].text
    csv_writer.writerow([Postcode, Borough, Neighbourhood])
csv_file.close()

df=pd.read_csv('toronto.csv')
```

```python
df[df.columns] = df.apply(lambda x: x.str.strip('\n'))
df = df[df.Borough!='Not assigned']
df.set_index(['Postcode','Borough'],inplace=True)
df = df.groupby(level=['Postcode','Borough'], sort=False).agg( ','.join)
df.dtypes

df.reset_index()
```

```python
df = df.groupby(['Postcode','Borough']).sum().reset_index()
df
```

1- Request page from Wikipedia
2- Used BeautifulSoup for find a Table inside a page
3- Used FOR loop to take information from every cell in a Table and put in a DataFrame (Pandas)
4- Renamed columns
5- Clear "Borough" column for "Not Assigned"
6- Group data from "Postcode" and "Borough"
